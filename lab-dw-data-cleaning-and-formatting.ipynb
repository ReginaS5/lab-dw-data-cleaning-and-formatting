## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
#Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".

data["gender"].unique()
```

```python
data["gender"] = data ["gender"].str.upper()
data["gender"].unique()
```

```python
def clean(x):
    if x in ["MALE","M"]:
        return "M"
    elif x in ["FEMAL", "FEMALE","F"]:
        return "F"
    else:
        return"U"
    
data["gender"] = list(map(clean,data["gender"]))
data["gender"].unique()
```

```python
data
```

```python
#State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"
data ['state'].value_counts()
```

```python
def clean(x):
    if x in ['California','Cali']:
        return 'California'
    elif x in ['Arizona','AZ']:
        return 'Arizona'
    elif x in ['Washington', 'WA']:
        return 'Washington'
    else:
        return x
    

data['state'] = list(map(clean, data['state']))
data['state'].unique()
data['state'].value_counts()
```

```python
#In education, "Bachelors" could be replaced by "Bachelor"
data['education'] = data['education'].replace('Bachelors','Bachelor')
```

```python
data
```

```python
#In Customer Lifetime Value, delete the % character

data['customer_life_time_value'] = data['customer_life_time_value'].str.replace('%', '')
```

```python
data
```

```python
#In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"
data['vehicle_class'].unique()  
print(data['vehicle_class'].value_counts())
def clean(x):
    if x in ['Sports Car', 'Luxury SUV','Luxury Car']:
        return 'Luxury'
    else:
        return x
    
data['vehicle_class'] = list(map(clean, data['vehicle_class']))
data['vehicle_class'].unique()  
data['vehicle_class'].value_counts()
```

```python
# Get the numeric columns
numerical = data.select_dtypes(include=['float64', 'int64'])

# Set up a grid of axes for plotting
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Flatten the axes for easy iteration
axes = axes.flatten()

# Loop through numerical columns and create distribution plots
for i, col in enumerate(numerical.columns):
    sns.histplot(data[col], kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Distribution of {col}')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = numerical.corr()

# Generate a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

```python
# Find pairs of features with high correlation (greater than 0.9)
high_correlation = (correlation_matrix > 0.9) & (correlation_matrix < 1)

# Get the indices of the features to drop
features_to_drop = []

for col in high_correlation.columns:
    correlated_cols = high_correlation.index[high_correlation[col]]
    for correlated_col in correlated_cols:
        if correlated_col != col and correlated_col not in features_to_drop:
            features_to_drop.append(correlated_col)

# Drop the features with high correlation
numerical_filtered = numerical.drop(columns=features_to_drop)

# Display the features that were dropped
print(f"Dropped features: {features_to_drop}")

# Display the updated correlation matrix
correlation_matrix_filtered = numerical_filtered.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_filtered, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numerical Features (Filtered)")
plt.show()
```
